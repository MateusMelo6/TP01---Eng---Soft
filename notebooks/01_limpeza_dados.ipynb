```python
import pandas as pd
```

```python
df = pd.read_csv('../data/INMET-2025-BH-PAMPULHA.CSV', encoding='latin-1', sep=';', skiprows=8, decimal=',')
```

```python
df.head()
```

```python
df = df.drop(columns=['Unnamed: 19'])
df.head()
```

```python
df.info()
```

```python
df['RADIACAO GLOBAL (Kj/m²)'].count()
```

```python
df.columns = [
    'data', 'hora_utc', 'precipitacao', 'pressao_atm', 'pressao_max', 'pressao_min',
    'radiacao_global', 'temp_ar', 'temp_orvalho', 'temp_max', 'temp_min',
    'temp_orvalho_max', 'temp_orvalho_min', 'umidade_max', 'umidade_min',
    'umidade_relativa', 'vento_direcao', 'vento_rajada', 'vento_velocidade'
]

df.head(2)
```

```python
# Limpar a coluna 'hora_utc' (de '0000 UTC' para '00:00:00')
df['hora_utc'] = df['hora_utc'].str[:2] + ':00:00'

# Criar uma coluna DateTime combinando data e hora
df['data_hora_utc'] = pd.to_datetime(df['data'] + ' ' + df['hora_utc'])

# Converter para o horário de Brasília (UTC-3)
df['data_hora_local'] = df['data_hora_utc'] - pd.Timedelta(hours=3)

# Definir a data_hora_local como o índice do DataFrame (ótimo para séries temporais)
df = df.set_index('data_hora_local')

# Excluir as colunas antigas que não precisamos mais
df = df.drop(columns=['data', 'hora_utc', 'data_hora_utc'])
```

```python
# 1. Zerar a radiação nos períodos noturnos/ausentes
df['radiacao_global'] = df['radiacao_global'].fillna(0.0)

# 2. Preencher pequenas falhas nos sensores (outras colunas) usando interpolação linear
cols_para_interpolar = df.columns.drop('radiacao_global')
df[cols_para_interpolar] = df[cols_para_interpolar].interpolate(method='linear')

# 3. Verificar se sobrou algum valor nulo
print(df.isnull().sum())
```

```python

df.to_csv('../data/INMET_Pampulha_Limpo.csv')
print("Dados limpos e salvos com sucesso!")
```
